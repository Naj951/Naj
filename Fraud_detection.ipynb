```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
#loading the data set to a pandas dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
#first 5 rows of the dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#data set information
credit_card_data.info()
```

```python
#checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
#distribution of legit transactions & fradulent transactions
credit_card_data['Class'].value_counts()
#0 normal
#1 fraud
```

```python
#seperatingthe data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the value for both transaction
credit_card_data.groupby('Class').mean()
```

```python
#undaer sampling
legit_sample = legit.sample(n=492)
#build a sample dataset containing similar distribution of normal transaction and fraudulent transaction
```

```python
# concatenating two data frames
new_dataset = pd.concat([legit_sample,fraud],axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#splitting data into features and target
```

```python
x= new_dataset.drop(columns= 'Class',axis = 1)
y= new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
#split data into training and testing data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#model training
#logistic regression
model = LogisticRegression(max_iter=1000)
#training the logistic regression model with training data
model.fit(x_train,y_train)
```

```python
#model evaluation
#accuracy score
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print('accuracy on training data :',training_data_accuracy)
```

```python
#acciracy on test data
x_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_prediction,y_test)
print('Accuracy score on test data:',test_data_accuracy)
```
